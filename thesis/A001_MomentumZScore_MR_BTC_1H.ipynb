### 2. The Rules (The "What")

Zscore:
<ul>
    <li>MA_Price = 20-period Simple Moving Average of Price</li>
    <li>StdDev_Price = 20-period Standard Deviation of Price</li>
    <li>Price_ZScore = (Price - MA_Price) / StdDev_Price</li>
</ul>
<br>
Momentum Zscore:
<ul>
    <li>ROC_1 = 1-period % Price Change</li>
    <li>MA_ROC = 50-period Simple Moving Average of ROC_1</li>
    <li>StdDev_ROC = 50-period Standard Deviation of ROC_1</li>
    <li>Momentum_ZScore = (ROC_1 - MA_ROC) / StdDev_ROC</li>
</ul>
<br><br>

When do I exit (Sell)?<br>
